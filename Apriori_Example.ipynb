```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori, association_rules
!pip install apyori
from apyori import apriori
```

```python
store_data = pd.read_excel('Buahan 2.xlsx', header=None)

store_data.head()
```

```python
num_records = len(store_data)
print(num_records)
```

```python
records = []
for i in range (0, num_records) :
    records.append([str(store_data.values[i,j]) for j in range(0,5)])
```

```python
association_rules = apriori(records, min_support=0.071, min_confidence = 0.2, min_lift = 3, min_length = 2)
association_results = list(association_rules)

for item in association_results :
     pair = item[0]
     items = [x for x in pair]
     print("Rule: " + items[0] + " --> " + items[1])



      
     print("Support : " + str(item[1]))




     print("Confidence : " + str(item[2][0][2]))
     print("Lift :" + str(item[2][0][3]))
     print("-----------------------")
print("Banyaknya Hasil Association Rules : " , len(association_results))
```
